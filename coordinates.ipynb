```python
import pandas as pd
import requests
import time

# Load your file
df = pd.read_csv("../data/locations.csv")

# Your LocationIQ token
TOKEN = "pk.87229b157a53d3ae8ee92c7149a8c667"

# Function to get coordinates
def get_coordinates(place):
    try:
        url = f"https://us1.locationiq.com/v1/search?key={TOKEN}&q={place}, Malaysia&format=json"
        response = requests.get(url)
        data = response.json()
        if isinstance(data, list) and len(data) > 0:
            return float(data[0]["lat"]), float(data[0]["lon"])
        else:
            return None, None
    except Exception as e:
        print(f"Error for {place}: {e}")
        return None, None

# Apply to all locations with delay to respect rate limit
coords = []
for place in df["Location"]:
    lat, lon = get_coordinates(place)
    coords.append((lat, lon))
    time.sleep(1)  # 1 second delay to stay within free-tier limits

# Add new columns
df["Latitude"], df["Longitude"] = zip(*coords)

# Save as coordinates.csv
df.to_csv("coordinates.csv", index=False)
print("✅ Done! File saved as coordinates.csv")
```
